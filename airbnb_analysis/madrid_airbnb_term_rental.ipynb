```python
calendar = pd.read_csv('./madrid_data/calendar.csv')

calendar['date']= pd.to_datetime(calendar['date'])

display(calendar.head(), calendar.describe())
```

### Seasonal Trends

```python
calendar['month']= calendar['date'].dt.month
calendar['season'] = calendar['month'].apply(lambda x: 'Winter' if x in [12, 1, 2] else ('Spring' if x in [3, 4, 5] else ('Summer' if x in [6, 7, 8] else 'Autumn')))
```

```python
calendar['price']=calendar['price'].replace('[\$,]','', regex=True).astype(float)
```

```python
# Average price per season
avg_price_season = calendar.groupby('season')['price'].mean().reset_index()   
avg_price_season.round(2)
```
